```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df_final=pd.merge(df,df_country,on='Country Code', how='left')
df_final.head()
```

```python
df_final.dtypes
```

```python
df_final.columns
```

```python
country_names=df_final.Country.value_counts().index
```

```python
country_val=df_final.Country.value_counts().values
```

```python
df_final.Country.value_counts().values
```

```python
## Pie Chart- Top 3 countries that uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

Observation:Zomato maximum records or transaction are from India After that USA and then United Kingdoms

```python
df_final.columns
```

```python
ratings=df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

# Observation
- 1.When Rating is between 4.5 to 4.9---> Excellent
- 2.When Rating are between 4.0 to 3.4--->very good
- 3.when Rating is between 3.5 to 3.9----> good
- 4.when Rating is between 3.0 to 3.4----> average
- 5.when Rating is between 2.5 to 2.9----> average
- 6.when Rating is between 2.0 to 2.4----> Poor

```python
ratings.head()
```

```python
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings)
plt.show()
```

```python
sns.barplot(x="Aggregate rating",y="Rating count",hue='Rating color',data=ratings,palette=['black','red','orange','yellow','green','green'])
plt.show()
```

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,palette=['black','red','orange','yellow','green','green'])
plt.show()
```
